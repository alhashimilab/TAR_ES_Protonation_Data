```python
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from scipy.optimize import curve_fit 
from uncertainties import unumpy as unp
import uncertainties as unc
```

```python
# Calculate Reduced Chi2 from fits:
def red_chi2(model, x_data, y_data, sigma, popt):
    # sigma = uncertainties in y
    
    # Residuals
    residuals = y_data - model(x_data, *popt)

    # Chi-squared
    chi2 = np.sum((residuals / sigma) ** 2)

    # Degrees of freedom
    dof = len(x_data) - len(popt)

    # Reduced chi-squared
    chi2_red = chi2 / dof
    
    return chi2_red
```

```python
# Fit and plot 5-state thermodynamic model for TAR
def model(x, pKa_ES2, Kconf_ES2):
    pKa_ES1 = 7.5
    Kconf_ES1 = 0.0126
    return Kconf_ES2*(1+10**-(x-pKa_ES2))/(1 + Kconf_ES1*(1+10**-(x-pKa_ES1)) + Kconf_ES2*(1+10**-(x-pKa_ES2)))

x_data = np.array([5.4, 6.0, 6.4, 7.0, 7.4])
pB = unp.uarray([1.697, 1.013, 0.40, 0.166, 0.497], [0.167, 0.142, 0.05, 0.013, 0.365])/100.0

#x_data = np.array([5.4, 6.0, 6.4, 7.0])
#pB = unp.uarray([1.697, 1.094, 0.40, 0.166], [0.167, 0.213, 0.05, 0.013])/100.0


y_data = unp.nominal_values(pB)

params, covariance = curve_fit(model, x_data[:-1], y_data[:-1], p0=[8.0, 0.01])

sigma = unp.std_devs(pB)
print("R-chi2 = ", red_chi2(model, x_data[:-1], y_data[:-1], sigma[:-1], params))

pKa_ES2 = params[0]
Kconf_ES2 = params[1]
pKa_ES2_err = np.sqrt(covariance[0][0])
Kconf_ES2_err = np.sqrt(covariance[1][1])
print(f"Fitted parameter: pKa(ES2) = {pKa_ES2} +/- {pKa_ES2_err}")
print(f"Fitted parameter: Kconf(ES2) = {Kconf_ES2} +/- {Kconf_ES2_err}")

fig, ax = plt.subplots(figsize=(4.5,3))
# Plot the results
plt.errorbar(x=x_data, y=y_data*100, yerr=unp.std_devs(pB)*100, label='NMR',
             marker='o', markersize=10, elinewidth=2, capsize=5, ecolor='k',
             capthick=2, color='k', ls="")

# y_data_ = unp.nominal_values(pB_)[-1:]
# plt.errorbar(x=x_data_[-1:], y=y_data_*100, yerr=unp.std_devs(pB_)[-1:]*100, label='NMR',
#              marker='o', markersize=10, elinewidth=2, capsize=5, ecolor='k',
#              capthick=2, color='k', ls="", mfc='none', markeredgewidth=2)


x_data = np.linspace(5.0, 8, 100)
plt.plot(x_data, 100.0*model(x_data, *params), color='tab:red', label='Fit curve')
plt.yticks([0.5, 1.0, 1.5, 2.0], fontsize=14)
plt.xticks(fontsize=14)
plt.ylabel('ES2 Population (%)', fontsize=16)
plt.xlabel('pH', fontsize=16)
plt.legend()
plt.xlim(5, 8)
plt.ylim(0, 2.5)
#plt.savefig("pKa_Fit_v2.pdf")
    
```
