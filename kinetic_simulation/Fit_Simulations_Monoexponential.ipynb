```python
import pandas as pd, numpy as np
import os, sys
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
#  Simulation for wt:
df = pd.read_csv("pH5.4//SimulationResults.csv")


# Fit data:
def func(x, Aeq, dA0, kex):
    return Aeq + dA0 * np.exp(-kex * x)

p0 = np.array([0.9, 0.1, 10000])
xdata = df['time(s)']
ydata = df['pop_state_3']

fig, ax = plt.subplots(1, figsize=(4,3))
ax.plot(xdata, ydata, label='Sim Data CS', ls="", marker='o', markersize=8, color='green')

popt, pcov = curve_fit(func, xdata, ydata, p0=p0)
ax.plot(xdata, func(xdata, *popt), 'k-', label='2-State Fit CS', lw=1.5)

ax.set_xlabel('Time (ms)')
ax.set_ylabel('Population', fontsize=16)

print("Fit kex = ", popt[-1])
print("Fit pB = ", 1-popt[0])
print("Fit kforward = ", popt[-1]*(1-popt[0]))
print("Fit kreverse = ", popt[-1] - popt[-1]*(1-popt[0]))

ax.legend()
fig.tight_layout()
#plt.savefig("CS_vs_IFRelax_wtSimulations.pdf")
plt.show()

```
